```python
# I'd like to make data adjustment in warehouse block, gender

df['warehouse_block'] = df['warehouse_block'].replace('F', 'E')
df['gender'] = df['gender'].replace({'F':'female', 'M':'male'})
df
```

```python
# i want to add non numerical columns to categorize customers based on its level of satisfaction derived from customer rating:

df['customer_rating_category'] = pd.cut(df['customer_rating'], bins=[0, 1, 2, 3, 4, 5], labels=['Very dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very Satisfied'])
df['on_time(str)'] = df['reached.on.time_y.n'].map({0: 'on time', 1: 'delayed'})
df


```

```python
# To identify the number of outliers present in the dataset, I will conduct an outlier analysis. 
# This analysis will help in understanding the extent of any extreme values that deviate significantly from the rest of the data.

numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
outlier_counts = {}

for col in numerical_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

outlier_counts
```

#### insight:

It turns out, there are outliers in 'prior_purchases' and 'discount_offered' but they will not be eliminated. Outliers in 'prior_purchases' and 'discount_offered' columns can provide valuable insights into customer behavior and preferences. Understanding why certain customers make more purchases or receive higher discounts can help businesses tailor in marketing strategies, loyalty programs, and pricing policies to better meet customer needs and improve overall performance.

```python
# Incorporating customer loyalty classification by categorizing them based on prior purchase history could offer valuable insights. 
# Therefore, creating additional columns to categorize customers based on their prior purchase behavior might be beneficial for understanding and analyzing customer loyalty trends.

df['prior_purchases'].value_counts()
```

```python
# making customer loyalty classification in to 3: low, moderate, high

def categorize_loyalty(prior_purchases):
    if prior_purchases < 4:
        return 'Low Loyal'
    elif prior_purchases <= 6:
        return 'Moderate Loyal'
    else:
        return 'High Loyal'

df['loyalty_category'] = df['prior_purchases'].apply(categorize_loyalty)
df
```

```python
# i want to add non numerical columns to categorize customers based on its level of satisfaction derived from customer rating:

df['customer_rating_category'] = pd.cut(df['customer_rating'], bins=[0, 1, 2, 3, 4, 5], labels=['Very dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very Satisfied'])
df['on_time(str)'] = df['reached.on.time_y.n'].map({0: 'on time', 1: 'delayed'})

```

```python
# To gain a clearer understanding of the performance of each warehouse block, I intend to create a comprehensive table where warehouses are grouped and analyzed in comparison with other relevant columns.


warehouse_group = df.groupby('warehouse_block').agg({'id': 'count',
                                                     'customer_care_calls':'mean',
                                                     'customer_rating':'mean',
                                                     'cost_of_shipping':'mean',
                                                     'prior_purchases':'mean',
                                                     'discount_offered(%)': 'mean',
                                                     'weight_in_gms': 'mean',
                                                     'reached.on.time_y.n': lambda x: (x == 0).mean() * 100
                                                     })
warehouse_group
```

```python
# Imagine that if the company has set its (Key Performance Indicator) KPI for every warehouse block. the KPI for on time rate, for instance 60%. Meanwhile, the customer rating KPI for each block is 3

# I define the KPI
on_time_delivery_kpi = 60
customer_rating_kpi = 3

# Plot for Customer Rating
plt.figure(figsize=(10, 6))

# Plot bars where warehouse is indexed, it is also grouped and given label
bars = plt.bar(warehouse_group.index, warehouse_group['avg_customer_rating'], color='indianred', label='Avg. Customer Rating')

# Plot Customer Rating KPI line
plt.axhline(y=customer_rating_kpi, color='blue', linestyle='--', label='Customer Rating KPI')

# Add value labels inside bars by looping
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

# Add labels and legend
plt.xlabel('Warehouse Block')
plt.ylabel('Avg. Customer Rating')
plt.title('Warehouse Performance - Customer Rating')
plt.legend()

# Show plot
plt.grid(axis='y')
plt.show()

# Plot another chart for On-Time Delivery Rate
plt.figure(figsize=(10, 6))

# Plot bars
bars = plt.bar(warehouse_group.index, warehouse_group['on_time_deliveries_success_rate'], color='lightblue', label='On-Time Delivery Rate (%)')

# Plot On-Time Delivery Rate KPI line with axhline
plt.axhline(y=on_time_delivery_kpi, color='red', linestyle='--', label='On-Time Delivery Rate KPI')

# Loop again to add value labels inside bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

# Add labels and legend
plt.xlabel('Warehouse Block')
plt.ylabel('On-Time Delivery Rate (%)')
plt.title('Warehouse Performance - On-Time Delivery Rate')
plt.legend()

# Show plot
plt.grid(axis='y')
plt.show()
```

```python
# Define a function to create comparison plots
def create_comparison_plot(data, x, y, title):
    plt.figure(figsize=(10, 6))
    sns.barplot(data=data, x=x, y=y, palette='inferno')
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel('Average Cost of Shipping')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Comparing average cost of shipping by mode of shipment
create_comparison_plot(df, 'mode_of_shipment', 'cost_of_shipping', 'Average Cost of Shipping by Mode of Shipment')

# Comparing average cost of shipping by customer satisfaction category
create_comparison_plot(df, 'customer_rating_category', 'cost_of_shipping', 'Average Cost of Shipping by Customer Satisfaction')

# Comparing average cost of shipping by loyalty category
create_comparison_plot(df, 'loyalty_category', 'cost_of_shipping', 'Average Cost of Shipping by Loyalty Category')

# Comparing average cost of shipping by product importance
create_comparison_plot(df, 'product_importance', 'cost_of_shipping', 'Average Cost of Shipping by Product Importance')

# Comparing average cost of shipping by warehouse block
create_comparison_plot(df, 'warehouse_block', 'cost_of_shipping', 'Average Cost of Shipping by Warehouse Block')
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Null Hypothesis (H0):
# There is no significant association between the mode of shipment and whether the shipment reached on time.

# Alternative Hypothesis (H1):
# There is a significant association between the mode of shipment and whether the shipment reached on time.

contingency_table = pd.crosstab(df['mode_of_shipment'], df['reached.on.time_y.n'])

# chi square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("P-value:", p)
```

##### Interpretation:

- With a p-value of 0.6895, which is greater than the typical significance level of 0.05, we fail to reject the null hypothesis.
- The chi-square statistic value of 0.7434 indicates the extent of the deviation of the observed frequencies from the expected frequencies under the null hypothesis.
A lower chi-square statistic suggests weaker evidence against the null hypothesis.
This indicates that there is insufficient evidence to conclude that there is a significant association between the mode of shipment and whether the shipment reached on time.

```python
contingency_table2 = pd.crosstab(df['warehouse_block'], df['reached.on.time_y.n'])

#Null Hypothesis (H0):
#There is no significant association between the warehouse block and whether the shipment reached on time.

#Alternative Hypothesis (H1):
#There is a significant association between the warehouse block and whether the shipment reached on time.
# chi square test

chi2, p, dof, expected = chi2_contingency(contingency_table2)

print("Chi-square statistic:", chi2)
print("P-value:", p)
```

- The chi-square statistic value of 1.089 indicates the extent of the deviation of the observed frequencies from the expected frequencies under the null hypothesis.
A lower chi-square statistic suggests weaker evidence against the null hypothesis.

This suggests that there is insufficient evidence to conclude that there is a significant association between the warehouse block and whether the shipment reached on time.
